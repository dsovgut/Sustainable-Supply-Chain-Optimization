## a) Calculate total costs and emissions of serving all regions from Denver

```python
from IPython.display import display, Markdown

# LaTeX formulas for cost and emissions
latex_formula = r"""
### Total Cost Calculation

1. **BSC Total Cost**:
$
\text{BSC Transport Cost} = \sum_{i=1}^5 (\text{Demand}_i \times \text{Transport Cost}_i)
$

$
\text{BSC Production Cost} = \text{BSC Variable Cost} \times \text{Total Demand for BSC}
$

$
\text{BSC Total Cost} = \text{BSC Fixed Cost} + \text{BSC Transport Cost} + \text{BSC Production Cost}
$

2. **CFP Total Cost**:
$
\text{CFP Transport Cost} = \sum_{i=1}^5 (\text{Demand}_i \times \text{Transport Cost}_i)
$

$
\text{CFP Production Cost} = \text{CFP Variable Cost} \times \text{Total Demand for CFP}
$

$
\text{CFP Total Cost} = \text{CFP Fixed Cost} + \text{CFP Transport Cost} + \text{CFP Production Cost}
$

3. **Total Cost for Both Items**:
$
\text{Total Cost} = \text{BSC Total Cost} + \text{CFP Total Cost}
$

---

### Total Emissions Calculation

1. **BSC Total Emissions**:
$
\text{BSC Transport Emissions} = \sum_{i=1}^5 (\text{Demand}_i \times \text{Transport Emissions}_i)
$

$
\text{BSC Production Emissions} = \text{BSC Emissions per Unit} \times \text{Total Demand for BSC}
$

$
\text{BSC Total Emissions} = \text{BSC Transport Emissions} + \text{BSC Production Emissions}
$

2. **CFP Total Emissions**:
$
\text{CFP Transport Emissions} = \sum_{i=1}^5 (\text{Demand}_i \times \text{Transport Emissions}_i)
$

$
\text{CFP Production Emissions} = \text{CFP Emissions per Unit} \times \text{Total Demand for CFP}
$

$
\text{CFP Total Emissions} = \text{CFP Transport Emissions} + \text{CFP Production Emissions}
$

3. **Total Emissions for Both Items**:
$
\text{Total Emissions} = \text{BSC Total Emissions} + \text{CFP Total Emissions}
$
"""

# Explanation of the process
explanation = r"""
**Process and Assumptions for Calculating Costs and Emissions:**

1. **Demand Distribution**:
   - The demand for both BSC and CFP items was distributed across five regions:
     - **Northeast, Southeast, Midwest, Southwest, Northwest.**

2. **Costs**:
   - Transportation costs were calculated by multiplying demand for each region by the transportation cost per item for that region.
   - Production costs were calculated by multiplying the total demand for each item by the respective variable cost per unit.
   - Total costs included fixed costs, transportation costs, and production costs.

3. **Emissions**:
   - Transportation emissions were calculated by multiplying demand for each region by the transport emissions per item for that region.
   - Production emissions were calculated by multiplying the total demand for each item by its emissions per unit.
   - Total emissions included both transportation and production emissions.

4. **Assumptions**:
   - All demand is met without exceeding facility capacity.
   - Costs and emissions scale linearly with demand and distance.
   - The selected cities accurately represent the regions geographically and economically.
   - The transportation mode (e.g., trucking) is consistent across all locations and distances.

"""

# Display the formulas and explanation
display(Markdown(latex_formula))
display(Markdown(explanation))
```

```python
from IPython.display import display, Markdown

# Corrected LaTeX write-up
write_up = r"""
## Optimization Model Explanation

### Objective Function
The goal is to minimize the total cost \( Z \), which comprises the following components:

1. **Fixed Costs**:
   $$ 
   \sum_{i \in F} F_i \cdot y_i 
   $$
   - \( F_i \): Fixed cost of opening facility \( i \).
   - \( y_i \): Binary decision variable, where \( y_i = 1 \) if facility \( i \) is open, and \( y_i = 0 \) otherwise.

2. **Production Costs**:
   $$ 
   \sum_{i \in F} \sum_{j \in R} \left( P_{BSC,i} \cdot x_{BSC,ij} + P_{CFP,i} \cdot x_{CFP,ij} \right) 
   $$
   - \( P_{BSC,i}, P_{CFP,i} \): Production cost per unit for BSC and CFP at facility \( i \).
   - \( x_{BSC,ij}, x_{CFP,ij} \): Number of units of BSC and CFP shipped from facility \( i \) to region \( j \).

3. **Transportation Costs**:
   $$ 
   \sum_{i \in F} \sum_{j \in R} \left( T_{ij} \cdot (x_{BSC,ij} + x_{CFP,ij}) \right) 
   $$
   - \( T_{ij} \): Transportation cost per unit from facility \( i \) to region \( j \).

4. **Emissions Costs**:
   $$ 
   \text{CarbonPrice} \cdot \left( 
   \sum_{i \in F} \sum_{j \in R} \left( E_{P,BSC,i} \cdot x_{BSC,ij} + E_{P,CFP,i} \cdot x_{CFP,ij} \right) + 
   \sum_{i \in F} \sum_{j \in R} \left( E_{T,ij} \cdot (x_{BSC,ij} + x_{CFP,ij}) \right)
   \right) 
   $$
   - \( \text{CarbonPrice} \): Internal carbon price per kilogram of CO\(_2\).
   - \( E_{P,BSC,i}, E_{P,CFP,i} \): Production emissions (kg CO\(_2\) per unit of BSC and CFP) at facility \( i \).
   - \( E_{T,ij} \): Transportation emissions (kg CO\(_2\) per unit shipped from facility \( i \) to region \( j \).

### Constraints
The objective function is subject to several constraints, including:

1. **Demand Satisfaction**:
   $$
   \sum_{i \in F} x_{BSC,ij} \geq D_{BSC,j}, \quad \forall j \in R 
   $$
   $$
   \sum_{i \in F} x_{CFP,ij} \geq D_{CFP,j}, \quad \forall j \in R 
   $$
   Ensures that the total units shipped meet or exceed regional demands.

2. **Capacity Limits**:
   $$
   \sum_{j \in R} x_{BSC,ij} \leq C_{BSC,i} \cdot y_i, \quad \forall i \in F 
   $$
   $$
   \sum_{j \in R} x_{CFP,ij} \leq C_{CFP,i} \cdot y_i, \quad \forall i \in F 
   $$
   Ensures that production does not exceed the available capacity at any facility.

3. **Facility Opening**:
   $$
   y_i \in \{0, 1\}, \quad \forall i \in F 
   $$

4. **Non-Negativity**:
   $$
   x_{BSC,ij}, x_{CFP,ij} \geq 0, \quad \forall i \in F, j \in R 
   $$

This function ensures a balance between cost minimization and sustainability, aligning with GreenPack's strategic goals.
"""

# Display the explanation as Markdown
display(Markdown(write_up))
```

```python
from pulp import LpProblem, LpMinimize, LpVariable, lpSum, LpBinary

# 1) Define sets
facilities = ['Denver', 'Charlotte', 'Columbus', 'Phoenix']
regions = ['Northeast', 'Southeast', 'Midwest', 'Southwest', 'Northwest']

# 2) Parameters (from your code snippet or re-arranged as dicts)

# Demand
demand_BSC = {
    'Northeast': 2_500_000,
    'Southeast': 2_000_000,
    'Midwest':   1_800_000,
    'Southwest': 1_200_000,
    'Northwest': 1_000_000
}
demand_CFP = {
    'Northeast': 1_500_000,
    'Southeast': 1_200_000,
    'Midwest':   1_000_000,
    'Southwest':   800_000,
    'Northwest':   500_000
}

# Facility capacities (BSC & CFP)
capacity_BSC = {
    'Denver':    8_000_000,
    'Charlotte': 4_000_000,  # If built
    'Columbus':  4_000_000,  # If built
    'Phoenix':   4_000_000   # If built
}
capacity_CFP = {
    'Denver':    5_000_000,
    'Charlotte': 3_000_000,
    'Columbus':  3_000_000,
    'Phoenix':   3_000_000
}

# Fixed costs for each facility's lines.
# If you only want "one binary open cost" per facility (both lines together),
# then you can combine them. Here we'll separate for clarity:
fixed_cost_BSC = {
    'Denver':    12_000_000,  # Already open, so you might skip or treat differently
    'Charlotte':  7_000_000,
    'Columbus':   7_000_000,
    'Phoenix':    7_000_000
}
fixed_cost_CFP = {
    'Denver':     8_000_000, 
    'Charlotte':  5_000_000,
    'Columbus':   5_000_000,
    'Phoenix':    5_000_000
}

# Variable production cost
prod_cost_BSC = {
    'Denver':    15,
    'Charlotte': 16,
    'Columbus':  16,
    'Phoenix':   16
}
prod_cost_CFP = {
    'Denver':    8,
    'Charlotte': 8.5,
    'Columbus':  8.5,
    'Phoenix':   8.5
}

# Production emissions (kg/unit)
prod_em_BSC = {
    'Denver':    2.0,
    'Charlotte': 1.8,
    'Columbus':  1.8,
    'Phoenix':   1.8
}
prod_em_CFP = {
    'Denver':    1.0,
    'Charlotte': 0.9,
    'Columbus':  0.9,
    'Phoenix':   0.9
}

# Transport costs ($/unit) and transport emissions (kg/unit)
transport_cost = {
    'Denver':    [8.5, 7.8, 5.2, 4.6, 4.2],
    'Charlotte': [3.027528, 2.738499, 3.918164, 5.880353, 8.909091],
    'Columbus':  [2.559551, 3.890557, 1.868263, 5.793451, 7.636364],
    'Phoenix':   [11.460674, 8.970944, 9.081836, 6.170025, 4.55]
}
transport_em = {
    'Denver':    [5.2, 4.8, 3.1, 2.8, 2.5],
    'Charlotte': [1.852135, 1.685230, 2.335828, 3.579345, 5.303030],
    'Columbus':  [1.565843, 2.394189, 1.113772, 3.526448, 4.545455],
    'Phoenix':   [7.011236, 5.520581, 5.414172, 3.755668, 2.708333]
}
# We'll map these by region index, so let's keep a region order list:
region_index = { 'Northeast': 0, 'Southeast': 1, 'Midwest': 2, 'Southwest': 3, 'Northwest': 4 }

# Carbon price
carbon_price = 0.10

# 3) Create the model
model = LpProblem("GreenPack_Optimization", LpMinimize)

# 4) Decision variables
#   xBSC[i,j] = quantity of BSC shipped from facility i to region j
#   xCFP[i,j] = quantity of CFP shipped from facility i to region j
#   yBSC[i]   = 1 if we open BSC line at facility i, 0 otherwise
#   yCFP[i]   = 1 if we open CFP line at facility i, 0 otherwise
#
#   If Denver is already open for both lines, we can fix them to 1 (or skip them).
xBSC = {}
xCFP = {}
for i in facilities:
    for j in regions:
        xBSC[(i,j)] = LpVariable(f"xBSC_{i}_{j}", lowBound=0, cat="Continuous")
        xCFP[(i,j)] = LpVariable(f"xCFP_{i}_{j}", lowBound=0, cat="Continuous")

# For new facilities, we have binary open/close. For Denver, we can either fix to 1 or just keep it as 1 in constraints.
yBSC = {}
yCFP = {}
for i in facilities:
    if i == "Denver":
        # If Denver is definitely open for both lines:
        yBSC[i] = LpVariable(f"yBSC_{i}", lowBound=1, upBound=1, cat="Continuous")  # fixed to 1
        yCFP[i] = LpVariable(f"yCFP_{i}", lowBound=1, upBound=1, cat="Continuous")  # fixed to 1
    else:
        yBSC[i] = LpVariable(f"yBSC_{i}", cat=LpBinary)
        yCFP[i] = LpVariable(f"yCFP_{i}", cat=LpBinary)

# 5) Objective function
#    Z = sum_i[ fixed(BSC,i)*yBSC[i] + fixed(CFP,i)*yCFP[i] ]
#         + sum_{i,j}[ (prod_cost_BSC[i] + transport_cost[i,j] + carbon_price * (prod_em_BSC[i]+transport_em[i,j])) * xBSC[i,j] ]
#         + sum_{i,j}[ (prod_cost_CFP[i] + transport_cost[i,j] + carbon_price * (prod_em_CFP[i]+transport_em[i,j])) * xCFP[i,j] ]
model += (
    lpSum(fixed_cost_BSC[i] * yBSC[i] for i in facilities) +
    lpSum(fixed_cost_CFP[i] * yCFP[i] for i in facilities) +
    lpSum(
        (prod_cost_BSC[i] 
         + transport_cost[i][region_index[j]] 
         + carbon_price * (prod_em_BSC[i] + transport_em[i][region_index[j]])) 
        * xBSC[(i,j)]
        for i in facilities for j in regions
    ) +
    lpSum(
        (prod_cost_CFP[i] 
         + transport_cost[i][region_index[j]] 
         + carbon_price * (prod_em_CFP[i] + transport_em[i][region_index[j]])) 
        * xCFP[(i,j)]
        for i in facilities for j in regions
    )
), "Total_Cost"

# 6) Constraints

# Demand must be met in each region for BSC and CFP
for j in regions:
    model += lpSum( xBSC[(i,j)] for i in facilities ) == demand_BSC[j], f"BSC_demand_{j}"
    model += lpSum( xCFP[(i,j)] for i in facilities ) == demand_CFP[j], f"CFP_demand_{j}"

# Capacity constraints for each facility
for i in facilities:
    # BSC capacity
    model += lpSum(xBSC[(i,j)] for j in regions) <= capacity_BSC[i] * yBSC[i], f"BSC_capacity_{i}"
    # CFP capacity
    model += lpSum(xCFP[(i,j)] for j in regions) <= capacity_CFP[i] * yCFP[i], f"CFP_capacity_{i}"

# If you want to force that a facility is either open for both or not at all, you can do:
# yBSC[i] == yCFP[i], but that is optional. 
# Or you can keep them separate to allow partial expansions.

# 7) Solve the model
model.solve()

print("Status:", model.status)
print("Objective value (Total Cost):", model.objective.value())

# 8) Print out the decisions
for i in facilities:
    print(f"\nFacility: {i}")
    print(f"  yBSC[{i}] = {yBSC[i].varValue}")
    print(f"  yCFP[{i}] = {yCFP[i].varValue}")
    for j in regions:
        flow_bsc = xBSC[(i,j)].varValue
        flow_cfp = xCFP[(i,j)].varValue
        if flow_bsc > 1e-5 or flow_cfp > 1e-5:
            print(f"    -> Region {j}: BSC={flow_bsc:,.0f}, CFP={flow_cfp:,.0f}")
```
